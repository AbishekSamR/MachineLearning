```python
df=pd.read_csv('GOOG.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df["Date"]=label_encoder.fit_transform(df["Date"])
df["Date"].unique()
```

```python
df
```

```python
x=df[["Date","Open","High","Low","Close","Adj Close",]]
x
y=df[["Volume"]]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
r2_score(ytest,ypred)

```

```python

prediction=model.predict([[1,2.515820,2.716817,2.503118,2.697639,2.697639]])
print(prediction)
```

```python
prediction=model.predict([[4857,	130.369995	,133.539993	,129.729996	,132.389999,	132.389999]])
print(prediction)
```

```python
prediction=model.predict([[4853	,138.985001	,139.669998	,136.294998	,136.399994	,136.399994]])
print(prediction)
```

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
model=linear_model.Lasso(alpha=0.3)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
score=r2_score(ytest,ypred)
print(score)
```

```python
prediction=model.predict([[1,2.515820,2.716817,2.503118,2.697639,2.697639]])
print(prediction)
```

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
model=linear_model.Ridge()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
score=r2_score(ytest,ypred)
print(score)
```

```python
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
score=r2_score(ytest,ypred)
print(score)
```

```python
prediction=model.predict([[0,	2.490664	,2.591785	,2.390042	,2.499133	,2.499133]])
print(prediction)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
score=accuracy_score(ytest,ypred)
score_1=r2_score(ytest,ypred)
print("Accuracy  :",score)
print("R2  :",score_1)
```
